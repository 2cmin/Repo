```python
#3-1(3-16)
%matplotlib inline
from sklearn.datasets import load_digits
digits = load_digits()

print("Image Data Shape", digits.data.shape)
print("Label Data Shape", digits.target.shape)
```

```python
#3-1(3-17)
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize = 20)
```

```python
#3-1(3-18)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=0)

from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)
```

```python
#3-1(3-19)
logisticRegr.predict(x_test[0].reshape(1,-1))
logisticRegr.predict(x_test[0:10])
```

```python
#3-1(3-20)
predictions = logisticRegr.predict(x_test)
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
#3-2
%matplotlib inline
from sklearn.datasets import load_digits
digits = load_digits()
print("Image Data Shape" , digits.data.shape)
print("Label Data Shape", digits.target.shape)

from PIL import Image
img1=Image.open('숫자5.png').convert("L")
img1=np.array(img1)
img1=img1/255.0
img1 = img1.flatten()

img2=Image.open('숫자6.png').convert("L")
img2=np.array(img2)
img2=img2/255.0
img2 = img2.flatten()

img3=Image.open('숫자7.png').convert("L")
img3=np.array(img3)
img3=img3/255.0
img3 = img3.flatten()

test_img = np.concatenate([[img1], [img2], [img3]])
print(test_img.shape)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.25, random_state=0)

from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)

logisticRegr.predict(x_test[0].reshape(1,-1))
logisticRegr.predict(x_test[0:10])

predictions = logisticRegr.predict(test_img)
print("예측값 : ", predictions)
label = np.array([5, 6, 7])
print("정답값 : ", label)
score = logisticRegr.score(test_img.reshape(3, -1), label)
print("정확도 : ", score)
```
